```python
#Initial set up
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

PROJECT_ROOT = "/content/neuroBrain"
DATASET_ROOT = "/content/Task01_BrainTumour"

IMAGES_PATH = os.path.join(DATASET_ROOT, "imagesTr")
LABELS_PATH = os.path.join(DATASET_ROOT, "labelsTr")

# Get sorted image and label lists
images = sorted(
    f for f in os.listdir(IMAGES_PATH)
    if f.endswith(".nii.gz") and not f.startswith("._")
)

labels = sorted(
    f for f in os.listdir(LABELS_PATH)
    if f.endswith(".nii.gz") and not f.startswith("._")
)

print(f"Number of MRI scans : {len(images)}")
print(f"Number of Labels    : {len(labels)}")
```

```python
#image-label correspondence verification which makes sure the filenames match
assert images[0] == labels[0], "Image and label filenames do not match!"

print("First image :", images[0])
print("First label :", labels[0])
print("\n Image-label correspondence verified.")
```

```python
#Loading one MRI and its mask
sample_image = os.path.join(IMAGES_PATH, images[0])
sample_label = os.path.join(LABELS_PATH, labels[0])

image = nib.load(sample_image)
label = nib.load(sample_label)

print(type(image))
print(type(label))
```

```python
#Inspecting metadata
print("=" * 15)
print("MRI Metadata")
print("=" * 15)

print("Shape          :", image.shape)
print("Voxel spacing  :", image.header.get_zooms())
print("Stored datatype:", image.get_data_dtype())
```

```python
#converting numpy data
image_np = image.get_fdata()
label_np = label.get_fdata()

print("Image shape :", image_np.shape)
print("Label shape :", label_np.shape)
```

```python
#checking the dimension match
assert image_np.shape[:3] == label_np.shape

print("Spatial dimensions match.")
```

## Observation

Unlike natural images, each MRI sample contains **four imaging modalities** (FLAIR, T1, T1ce, and T2), resulting in a 4D tensor of shape `(240, 240, 155, 4)`.

The segmentation mask is a single 3D volume of shape `(240, 240, 155)` where each voxel stores the corresponding tumor class label.

The spatial dimensions of the MRI and segmentation mask are identical, ensuring voxel-wise correspondence.


```python
# Find slices that contain tumor

tumor_slices = np.where(label_np.sum(axis=(0, 1)) > 0)[0]

print("Number of slices containing tumor:", len(tumor_slices))
print("First tumor slice:", tumor_slices[0])
print("Last tumor slice:", tumor_slices[-1])
```

```python
# Picking a representative slide
slice_idx = tumor_slices[len(tumor_slices) // 2]

print("Selected slice:", slice_idx)
```

```python
# Visualizing all four modalities
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

titles = ["FLAIR", "T1", "T1ce", "T2"]

for i in range(4):
    axes[i].imshow(image_np[:, :, slice_idx, i], cmap="gray")
    axes[i].set_title(titles[i], fontsize=14)
    axes[i].axis("off")

plt.suptitle(f"MRI Modalities (Slice {slice_idx})", fontsize=16)

plt.show()
```

```python
# Visualizing segmentation mask
plt.figure(figsize=(6, 6))

plt.imshow(label_np[:, :, slice_idx], cmap="viridis")

plt.title("Ground Truth Segmentation Mask")

plt.axis("off")

plt.show()

print("Unique labels:", np.unique(label_np))
```

## Initial Observation

So we found T1ce and T2 modalities are better in highlighting the tumor than other modalities ,among those T1ce looks best for it ,</br>

T1 emphasizes anatomy because we can clearly distinguish Gray matter ,White matter and Brain boundaries </br>

Unique label values present are `[0. 1. 2. 3.]` </br>

We aslo found appearance differences between modalities </br>

## Final Observations

- Each patient consists of four MRI modalities: FLAIR, T1, T1ce, and T2.
- The modalities provide complementary information about the brain.
- FLAIR and T2 highlight the tumor and surrounding edema more prominently.
- T1 provides clearer visualization of normal brain anatomy but lower tumor contrast.
- T1ce highlights the enhancing boundary of the tumor after contrast administration.
- The segmentation mask contains four classes (0–3), including the background and three tumor-related regions.
- Tumor tissue is present in 74 of the 155 axial slices for this patient.

So we found that No single MRI modality provides all the information needed to characterize the tumor.



```python
#Visualization function
def show_modalities(image_np, label_np, slice_idx):
    """
    Display all MRI modalities, segmentation mask,
    and overlay for a selected slice.
    """

    titles = ["FLAIR", "T1", "T1ce", "T2"]

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # First row
    for i in range(3):
        axes[0, i].imshow(image_np[:, :, slice_idx, i], cmap="gray")
        axes[0, i].set_title(titles[i], fontsize=13)
        axes[0, i].axis("off")

    # Second row
    axes[1, 0].imshow(image_np[:, :, slice_idx, 3], cmap="gray")
    axes[1, 0].set_title("T2", fontsize=13)
    axes[1, 0].axis("off")

    axes[1, 1].imshow(label_np[:, :, slice_idx], cmap="viridis")
    axes[1, 1].set_title("Ground Truth Mask", fontsize=13)
    axes[1, 1].axis("off")

    # Overlay
    axes[1, 2].imshow(image_np[:, :, slice_idx, 0], cmap="gray")
    axes[1, 2].imshow(
        label_np[:, :, slice_idx],
        cmap="jet",
        alpha=0.45
    )

    axes[1, 2].set_title("Mask Overlay (FLAIR)", fontsize=13)
    axes[1, 2].axis("off")

    plt.suptitle(
        f"Brain MRI Modalities and Segmentation (Slice {slice_idx})",
        fontsize=18
    )

    plt.tight_layout()

    return fig
```

```python
#Generating the figure
fig = show_modalities(image_np, label_np, slice_idx)
```

```python
#Saving the figure
import os

FIGURES_DIR = os.path.join(PROJECT_ROOT, "figures")

os.makedirs(FIGURES_DIR, exist_ok=True)

fig.savefig(
    os.path.join(FIGURES_DIR, "01_modalities_overview.png"),
    dpi=300,
    bbox_inches="tight"
)

print("Figure saved successfully!")
```
